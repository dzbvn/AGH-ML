```python
import tensorflow as tf
(X_train, y_train), (X_test, y_test) = tf.keras.datasets.boston_housing.load_data()
```

```python
from tensorflow import keras
from keras import optimizers

def build_model(n_hidden, n_neurons, optimizer, learning_rate, momentum=0): 
    model = tf.keras.models.Sequential()
    for i in range(n_hidden):
        model.add(keras.layers.Dense(n_neurons))
    if optimizer == 'sgd':
        model.compile(optimizer=keras.optimizers.SGD(learning_rate=learning_rate), loss=keras.losses.MeanSquaredError(), metrics=[
        keras.metrics.MeanAbsoluteError()
    ])
    elif optimizer == 'nesterov':
        model.compile(optimizer=keras.optimizers.SGD(learning_rate=learning_rate, nesterov=True), loss=keras.losses.MeanSquaredError(), metrics=[
        keras.metrics.MeanAbsoluteError()
    ])
    elif optimizer == 'momentum':
        model.compile(optimizer=keras.optimizers.SGD(learning_rate=learning_rate, momentum=momentum), loss=keras.losses.MeanSquaredError(), metrics=[
        keras.metrics.MeanAbsoluteError()
    ])
    elif optimizer == 'adam':
        model.compile(optimizer=keras.optimizers.Adam(learning_rate=learning_rate), loss=keras.losses.MeanSquaredError(), metrics=[
        keras.metrics.MeanAbsoluteError()
    ])
    return model
```

```python
es = tf.keras.callbacks.EarlyStopping(patience=10,
                                      min_delta=1, 
                                      verbose=1)
```

```python
for lr in [1e-4, 1e-5, 1e-6]:
    model1 = build_model(1, 25, 'sgd', lr)
    history = model1.fit(X_train, y_train, epochs=100, validation_split=0.1, callbacks=[es])
    print((lr, history.history['loss'][-1], history.history['mean_absolute_error'][-1]))
```

```python
for hl in [0, 1, 2, 3]:
    model1 = build_model(hl, 25, 'sgd', 1e-5)
    history = model1.fit(X_train, y_train, epochs=100, validation_split=0.1, callbacks=[es])
```

```python
for nn in [5, 25, 125]:
    model1 = build_model(1, nn, 'sgd', 1e-5)
    history = model1.fit(X_train, y_train, epochs=100, validation_split=0.1, callbacks=[es])
```

```python
for opt in ['sgd', 'nesterov', 'momentum', 'adam']:
    model1 = build_model(1, 25, opt, 1e-5)
    history = model1.fit(X_train, y_train, epochs=100, validation_split=0.1, callbacks=[es])
```

```python
for mom in [0.1, 0.5, 0.9]:
    model1 = build_model(1, 25, 'momentum', 1e-5, mom)
    history = model1.fit(X_train, y_train, epochs=100, validation_split=0.1, callbacks=[es])
```
